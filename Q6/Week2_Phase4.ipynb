Phase 4:Statistical Analysis and Probability

```python
import pandas as pd
import numpy as np
print("Data Statistics:)")
data=[10,20,30,40,50,60]
series=pd.Series(data)
mode=series.mode()
mean=series.mean()
median=series.median()
std=series.std()
var=series.var()
min=series.min()
max=series.max()
range=max-min
print("Range:",range)
print("Minimum:",min)
print("Maximum:",max)
print("Variance:",var)
print("Standard Deviation:",std)
print("Mode:",mode)
print("Mean:",mean)
print("Median:",median)
```

Binomial Distribution

```python
print("Binomial Distribution:)")
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
# Normal Distribution
x_norm = np.linspace(-5, 5, 1000)
y_norm = stats.norm.pdf(x_norm, loc=0, scale=1)
# Binomial Distribution
n = 20      # number of trials
p = 0.5     # probability of success
x_binom = np.arange(0, n+1)
y_binom = stats.binom.pmf(x_binom, n, p)
plt.figure(figsize=(10, 5))
plt.plot(x_norm, y_norm, label='Normal Distribution', color='blue')
plt.stem(x_binom, y_binom, linefmt='r-', markerfmt='ro', basefmt=" ", label='Binomial Distribution')
plt.title("Normal vs Binomial Distribution")
plt.xlabel("x")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Hypothesis_Testing

```python
print("Hypothesis Testing:)")
from scipy.stats import ttest_1samp
sample = [10, 12, 14, 13, 15]
t_stat, p_value = ttest_1samp(sample, 12)
print("p-value:", p_value)
print("t-statistic:", t_stat)
scores=[78, 80, 77, 76, 79]
t_stat,p_value=ttest_1samp(scores,75)
print("t-statistic of score:",t_stat)
print("p-value of score:",p_value)
if p_value<0.05 and t_stat>0:
    print("The average score is significantly greater than 75.")
else:
    print("No significant evidence that the average score is greater than 75.")
```

Confident Intervals and p_values

```python
print("Confidence Intervals and p-values:)")
import scipy.stats as stats
import numpy as np
Sales_sample=[220,230,250,560,90]
confident_interval=stats.t.interval(0.95,len(Sales_sample)-1,loc=np.mean(Sales_sample),scale=stats.sem(Sales_sample))
print("95% Confidence Interval:",confident_interval)
```

Central Limit Theorem

```python
print("Central Limit Theorem:)")
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
for _ in range(1000):
  sample_space=[np.mean(np.random.randint(0,100,50))]
print("Sample_Space: ",sample_space)
plt.hist(sample_space,bins=50,color='skyblue',edgecolor='black')
plt.title("CLT Approximation with Sample Size = 50 (1000 Samples)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

A_B Testing

```python
print("A/B Testing:)")
from scipy.stats import ttest_ind
A = [30, 35, 40, 38, 33]
B = [32, 34, 37, 39, 35]
t_stat, p = ttest_ind(A,B)
print("p-value:", p)
print("t-statistic:", t_stat)
if p < 0.05:
    print("There is a significant difference between the means of A and B.")
else:
    print("There is no significant difference between the means of A and B.")
```
